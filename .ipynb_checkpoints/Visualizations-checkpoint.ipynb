## Target Variable

#### Arrest Flag

```python
# how many traffic stops ended in arrest
arrest = df['Arrest_Flag']=='Y'
no_arrest = df['Arrest_Flag']=='N'

y = df['Arrest_Flag']
```

```python
num_arrests = df[arrest].shape[0]
num_noarrest = df[no_arrest].shape[0]
```

```python
# Arrest count
print('Target Variable')
print('Total Arrests: {}\nTotal Non-Arrests: {}'.format(num_arrests, num_noarrest))
print('Model Null Probability of being Arrested: {}'.format(round(num_arrests/(num_arrests + num_noarrest), 3)))

# Target Variable Countplot
sns.set_style('darkgrid');
plt.figure(figsize = (10,5));
sns.countplot(df['Arrest_Flag'], alpha =.80, palette= ['grey','lightgreen'])
plt.title('Arrest vs Non-Arrest');
plt.ylabel('# Subjects');
plt.show()
```

## Feature Engineering

Based on the first stacked histogram, there appears to be a better chance of being arrested by young officers, perhaps those who recently joined the force and are either looking to prove themselves or have quotas to fill.

```python
df['Young_Officer'] = np.where(df['Officer_Age'] <= 29, 1, 0)
```

Based on the second stacked histogram, it would appear that while there are more terry stops in the late-night hours, arrests stays relatively constant.

```python
df['Late_Night'] = np.where((df['Reported_Hour']>=20) | (df['Reported_Hour']<=4), 1, 0)
```
